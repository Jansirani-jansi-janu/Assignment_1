```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('sqlite:///food_wastage.db')

# Load CSV files
providers = pd.read_csv('/content/drive/MyDrive/LFWMS/providers_data.csv')
receivers = pd.read_csv('/content/drive/MyDrive/LFWMS/receivers_data.csv')
food_listings = pd.read_csv('/content/drive/MyDrive/LFWMS/food_listings_data.csv')
claims = pd.read_csv('/content/drive/MyDrive/LFWMS/claims_data.csv')

# Insert data into the database
providers.to_sql('providers', con=engine, if_exists='replace', index=False)
receivers.to_sql('receivers', con=engine, if_exists='replace', index=False)
food_listings.to_sql('food_listings', con=engine, if_exists='replace', index=False)
claims.to_sql('claims', con=engine, if_exists='replace', index=False)
```

```python
df = pd.read_sql('select * from providers', con=engine)
print(df.head())
```

```python
query = """
SELECT providers.city,
       COUNT(DISTINCT providers.provider_id) AS num_providers,
       COUNT(DISTINCT receivers.receiver_id) AS num_receivers
FROM providers
JOIN receivers ON providers.city = receivers.city
GROUP BY providers.city
"""
pd.read_sql_query(query, conn)
```
